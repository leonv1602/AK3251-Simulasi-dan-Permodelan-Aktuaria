## Apa itu Model Probabilistik

Model ini dimanfaatkan jika situasi deterministik tidak mencukupi atau tidak relevan. Dengan situasi yang makin kompleks, keadaan tersebut makin banyak dijumpai. Sekarang dengan kemampuan komputer yang luar biasa, mencoba mensimulasikan sistem dengan derajat besar, dan prosedur pemodelan klasik makin ditinggalkan.

### Contoh
1. Katakanlah dalam situasi fisis yang sudah dapat ditentukan dengan menggunakan hukum alam, tetapi kemudian akan ada parameter yang sulit ditentukan nilainya dengan keakurasian yang diinginkan. 

Misalkan saja, industri minyak ingin melakukan pengukuran geologi dan retakan yang ada di dalam bumi. 

2. Situasi fisis  yang dapat di modelkan dengan hukum alam yang ada, tetapi terjadi dengan perubahan yang sangat cepat dan tidak dapat terduga gaya luar. 

Misalkan saja pegas di mobil yang rusak akibat lubang jalan yang tidak menentu.

3. Situasi fisis yang tidak dapat dimodelkan dengan hukum alam yang  telah diketahui. 

Misalkan saja, tumbukan partikel atau atom, turbulensi dari fluida. Juga mengenai harga saham, waktu hidup orang dan lain sebagainya.

## Apakah bilangan random dapat menyelesaikan masalah

Simulasi atau khususnya model Simulasi adalah teknik untuk menganalisa fenomena dengan menggunakan algoritma komputer yang secara prinsip memanfaatkan bilangan random. Nama Monte Carlo diberikan oleh Stanislaw Ulam dan John von Neumann, yang menemukan metoda untuk menyelesaikan masalah difusi neutron di Los Alamos pada tahun 1940.
Nama Monte Carlo, setelah sebuah casino Monaco, mensimulasi kejadian probabilistik yang rumit, hanya dengan menggunakan kejadian yang lebih sederhana, misalkan dengan menggunakan dadu. 
Saat sekarang, kita akan menggunakan pseudorandom yang terletak di interval $[0,1]$ dan kemudian digunakan untuk membangun data dengan distribusi tertentu.

Sekarang, metode Monte Carlo ini dipakai sangat luas  di matematika, sains, industri, perdagangan dan bahkan hiburan. Dengan menggunakan algoritma komputer untuk membuat prediksi dari suatu proses stokastik, atau fenomena yang mempunyai komponen random.

### Masalah Jarum Buffon

Misalkan kita akan melempar $n$ jarum dengan panjang $L$ pada suatu daerah datar yang dilengkapi dengan garis-garis yang berjarak $d$. Misalkan $H_n$ adalah variabel random yang menyatakan banyak jarum yang menyentuh atau melintasi garis jika ada $n$ jarum yang dilempar.

https://mste.illinois.edu/activity/buffon/

Misalkan $X$ menyatakan jarak titik tengah jarum yang telah jatuh di lantai terhadap garis yang paling dekat dengan jarum. Misalkan pula $\theta$ menyatakan sudut lancip jarum terhadap garis. Dalam hal ini nilai $X$ adalah
$$
0\leq x \leq \frac{d}{2} \qquad \mbox{dan} \qquad 0 \leq \theta \leq \frac{\pi}{2}.
$$
Jarum ini akan memotong atau menyentuh garis jika dan hanya jika 
$$
X \leq \frac{L}{2} \sin \theta
$$

```python
from numpy import *
from numpy.random import  *
import matplotlib.pyplot as plt
lempar=1000000 # ada sebanyak lempar jarum
x=rand(lempar)
theta=rand(lempar)
theta=0.5*pi*theta
kena=[1 for i in range(lempar) if x[i]<=0.5*sin(theta[i])]
peluang_exp=sum(kena)/lempar
print(peluang_exp)
```

Peluang tersebut dapat dihitung melalui
$$
\frac{\mbox{daerah di bawah lengkungan}}{\mbox{luas kotak $(0,\pi/2)\times (0,d/2)$}}=\frac{2L}{\pi d}
$$

```python
L=1
djarak=2
peluang=2*L/(pi*djarak)
print(peluang)
```

Berdasarkan hasil tersebut, kita dapat memperkirakan nilai $\pi$ yaitu:

```python
Nilai_pi=2*L/(djarak*peluang_exp)
print(Nilai_pi)
```
