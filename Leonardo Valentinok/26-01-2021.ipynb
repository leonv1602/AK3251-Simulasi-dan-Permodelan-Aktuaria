```python
import numpy as np
np.set_printoptions(formatter = {'float' : '{:5.4f}'.format})
import random
import matplotlib.pyplot as plt
```

```python
a = 3
m = 2**3 - 1
v = 2
x = [v]
n = 100
for i in range(n):
    v = np.mod(a*v,m)
    x.append(v)
plt.hist(x, bins = 6)
```
